## 자율주행 분야 

### 구글 웨이모

- 주요 자율주행자동차 업체들 중 웨이모, GM 크루즈, 애플, 우버 등에서 웨이모가 가장 돋보인다.
- 자율 주행 해제 건수는 주행거리에 비해서 가장 적은 횟수를 보여주며 평균 자율주행 거리 또한 나머지 3곳의 기업을 제외하고도 가장 높다.
- 자율 주행은 단계 즉 Level로 나뉘어지며 Level이 높아질수록 고도화된 자율 주행 기술을 의미하며 웨이모는 Level4에 해당한다.
- 자율 주행의 가장 큰 장점은 사람들이 살아오던 행동 패턴에 대해 긍정적인 영향과 함께 혁신을 가져다주게 된다.
- Level5의 시대가 들어서게 되면 운전석에 대한 개념 자체가 사라지며 자동차로 이동하는 시간에도 원하는 곳에 시간을 할당할 수 있다.
- 아직은 자율주행차에 대한 신뢰도가 높지 않으며 사고 또한 자주 일어나는 추세이기에 자율주행기술의 부족함이 많다.
- CES 2020에서 현대와 우버가 함께 선보인 개인용 비행체처럼 미래에는 자율주행이 차량이 아닌 비행체까지 대체될 가능성이 높다.
